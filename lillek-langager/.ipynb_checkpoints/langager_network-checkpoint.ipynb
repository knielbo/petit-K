# Exploration and Testing of Temporal Observational Data from KWK #

__Hypothesis__:

H1: Reducation in behavioral repertoire increase probablility of success <span style="color:red">*Rejected*</span>.  
H2: Sequential non-attendance decrease probability of success <span style="color:red">*Rejected*</span>.  
H3: Significant change point 05-01-22, pre to post increase in probablility of success  <span style="color:red">*Rejected*</span>.

In conclusion, H1-H3 finds no support in the data, the exploratory data analysis provides an explanation by showing that success throughout the observation period is a white noise process and that observable success above the 95% and 99% confidence level are statistical flukes. Importantly, the analysis also explains why one might be led to the mistaken conclusion of a $success \times total$ correlation by analyzing relative success. This however is an artefact for modifying the numerator of relative success systematically.

## H3: Significant change point 05-01-22, pre to post increase in probablility of success ##

```python
x1 = df['total-unit'].iloc[:53].tolist()
y1 = df['success-unit'].iloc[:53].tolist()
x2 = df['total-unit'].iloc[53:].tolist()
y2 = df['success-unit'].iloc[53:].tolist()

print('pre intervention...')
x1 = sm.add_constant(x1)
result1 = sm.OLS(y1, x1).fit()
print(result1.summary())

print('post intervention...')
x2 = sm.add_constant(x2)
result2 = sm.OLS(y2, x2).fit()
print(result2.summary())
```

_in conclusion_ no support for a reliable effect of date in atemporal linear model

# Exploratory Data Analysis #
## Relative Success is an Error-prone Model ##

A mistaken account is relative success, because any absolute effect is just a function of noise (see analysis 1). Relative Success is effectively 'spreadsheet magic', you are just narrowing the band around a random process!!!

```python
x = df['idx01'].tolist()
y1 = df['success-unit'].tolist()
y2 = df['total-unit'].tolist()
```

```python
Y = list()
for i, val in enumerate(x):
    y_hat = y1[i]/y2[i]
    Y.append(y_hat)
df['y_hat'] = Y
```

```python
ax = sns.regplot(x='idx01', y='y_hat', data=df,
                 scatter_kws={"s": 25},
                 order=1, ci=None)
```

to show why this is erronous, let us compare absolute success on time with the number of tasks

```python
fig, ax = plt.subplots()
#ax= sns.lineplot(x='idx01', y='success-unit', data=df)

ax = sns.regplot(x='idx01', y='success-unit', data=df,
                 scatter_kws={"s": 25},
                 order=1, ci=None)
ax1 = sns.lineplot(x='idx01', y='total-unit', data=df)
```

in other words, you are only modifying the denominator, the numerator represents points in a random process.

## Is the treatment a random walk ##

THe time series is probably not white noise if one or more of the following conditions are true:

1. Is the mean/level non-zero?
2. Does the mean/level change over time?
3. Does the variance change over time?
4. Do values correlate with lag values?

Some tools that you can use to check if your time series is white noise are:

1. Create a line plot. Check for gross features like a changing mean, variance, or obvious relationship between lagged variables.
2. Calculate summary statistics. Check the mean and variance of the whole series against the mean and variance of meaningful contiguous blocks of values in the series (e.g. days, months, or years).
3. Create an autocorrelation plot. Check for gross correlation between lagged variables.
